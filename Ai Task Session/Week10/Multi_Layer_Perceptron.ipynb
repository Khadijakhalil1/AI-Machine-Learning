```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from sklearn import datasets
import matplotlib.pyplot as plt
```

**Classification model**

```python
iris = datasets.load_iris()
```

```python
iris.target
```

```python
x= iris.data
```

```python
y = iris.target
```

```python
iris.feature_names
```

```python
model = Sequential()
model.add(Dense(4,activation = 'relu',input_dim = 4)) # Input Layer
model.add(Dense(3,activation = 'softmax'))# output Layer
```

```python
model.summary()
```

```python
model.get_weights() # keras ne jo randomly weights aur biases initialize kye hon gy
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer=optimizer,metrics=['accuracy']) # agr categorical cross entropy use krna hai to .... data ko one hot encode krna ho ga
```

```python
history = model.fit(x,y,epochs = 75,verbose = 1, batch_size = 1, validation_split=0.2 )
```

**Loss Plot**

```python

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model Loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(['Train', 'Validation'])
plt.show()
```

Validation aur training loss mein bht zada gap hai mtlb model overfit kr rha hai

**Accuracy Plot**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(['Train', 'Validation'])
plt.show()
```

Validation Accuracy is low and unstable --> Sign of overfitting

**Regression Problem**

```python
housing = datasets.fetch_california_housing()
```

```python
housing.target
```

```python
x = housing.data
```

```python
y = housing.target
```

```python
housing.feature_names
```

```python
model = Sequential()
model.add(Dense(8,activation = 'linear',input_dim = 8)) # Input Layer
model.add(Dense(1,activation = 'linear'))# output Layer
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.001)
```

```python

model.compile(loss='mean_squared_error',optimizer=optimizer,metrics=['accuracy'])
```

```python
history = model.fit(x,y,epochs = 75,verbose = 1, batch_size = 1, validation_split=0.2 )
```

**Loss Plot**

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model Loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(['Train', 'Validation'])
plt.show()
```

validation loss is too high ....... it means overfitting occurs

**Accuracy Plot**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(['Train', 'Validation'])
plt.show()
```
